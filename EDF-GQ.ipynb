```python
import numpy as np
```

```python
class Task:
    ## Class for a task
    def __init__(self,index,c,t,d,i):
        self.index = index
        self.comp = c
        self.period = t
        self.deadline = d
        self.arrival = i
        
        #Keep track of remaining execution in this instance
        self.exec = c
        
    def out(self):
        print("Task no:"+str(self.index))
        print("("+str(self.comp)+","+str(self.period)+","+str(self.deadline)+","+str(self.arrival)+")")
```

```python
class Processor:
    def __init__(self,index):
        self.index = index
        self.tasks = set()        
        
    def accomodate(self,task):
        dbf = 0
        utilization = 1
        for sched in self.tasks:
            util = sched.comp/sched.period
            dbf += sched.comp + (task.deadline-sched.deadline)*util
            utilization -= util
        if utilization < (task.comp/task.period) or (task.deadline-dbf < task.comp):
            return False
        self.tasks.add(task)
        return True
    
    def schedule(self,time):
        sched = None
        for task in self.tasks:
            if task.arrival <= time:
                if sched == None:
                    sched = task
                elif task.deadline < sched.deadline:
                    sched = task
        if sched == None:
            return None
        sched.exec -=1
        if sched.exec == 0:
            sched.exec += sched.comp
            sched.arrival += sched.period
            sched.deadline += sched.period
        return sched
```

```python
# Partitioning 


filename = "tasks_multicore"
file = open(filename,"r")
lines = file.readlines()
n = int(lines[0])
if n < 1:
    print("Invalid number of tasks")
p = int(lines[1])    
    
tasks = [None for _ in range(n)]
lcm = 1
for i in range(2,len(lines)):
    if tasks[-1] != None :
        print("More than "+str(n)+" tasks")
        exit()
    vals = [int(el) for el in lines[i].split(',')]
    tasks[i-2] = Task(i-1,vals[0],vals[1],vals[2],vals[3])
    lcm = np.lcm(lcm,tasks[i-2].period)
if tasks[-1] == None:
    print("Less than "+str(n)+" tasks")

tasks = sorted(tasks,key = lambda x : x.period)    
    
for task in tasks :
    task.out()
print(lcm)
```

```python
processors = []
for i in range(p):
    processors.append(Processor(i+1))

for task in tasks:
    accom = False
    for i in range(len(processors)):
        if processors[i].accomodate(task) == True:
            print("Processor "+str(processors[i].index)+" accomodates task "+str(task.index))
            accom = True
            break
    if accom == False:
        print("Task "+str(task.index)+" not accomodated")
        break
```

```python
for time in range(lcm):
    print("Time: ",time)
    for i in range(len(processors)) :
        
        line = "Processor "+str(i+1)+" :"
        sched = processors[i].schedule(time)
        if sched == None:
            line +="Idle"
        else:
            line += "Task " +str(sched.index)
        print(line)
    print()
```
